# Figure 2: BIC curves

This notebook will reproduce Figure 2 from *Heuristic Methods for Determining the Number of Classes in Unsupervised Classification of Climate Models*, E. Boland et al. 2022 (doi to follow). This requires the file model_19651995/BICs2-30.obj, produced using calculate_BICcurves.ipynb

```python
import numpy as np
import matplotlib.pyplot as plt

import os
import pickle
```

```python
ids = ['r1i1p1f2', 'r2i1p1f2', 'r3i1p1f2'] # Ensemble members
#model_folder='model_19651995'
model_folder='model_20012017'
```

### Load data

```python
with open('{}/BICs2-20.obj'.format(model_folder), 'rb') as file:
    BICs = pickle.load(file)
with open('{}/AICs2-20.obj'.format(model_folder), 'rb') as file:
    AICs = pickle.load(file)
with open('{}/SILs2-20.obj'.format(model_folder), 'rb') as file:
    SILs = pickle.load(file)
```

### Plot BIC and $\Delta$BIC curves

```python
fig, axes = plt.subplots(1,2)
for mm,m_id in enumerate(ids):
    axes[0].plot(np.arange(2, 21), BICs[m_id], label=m_id)
    axes[1].plot(np.arange(3, 21), BICs[m_id][1:]-BICs[m_id][:-1], label=m_id)
fig.set_size_inches(12,6)
for ax in axes:
    ax.set_xlabel('Number of classes (K)')
    ax.set_xticks(np.arange(2,21,2))
axes[0].legend()
axes[0].set_title('BIC',fontsize=14,fontweight='bold')
axes[0].text(-0.08,1.0,'a)',transform=axes[0].transAxes,fontsize=16,fontweight='bold')
axes[1].set_title('$\Delta$BIC',fontsize=14,fontweight='bold')
axes[1].text(-0.08,1.0,'b)',transform=axes[1].transAxes,fontsize=16,fontweight='bold')
plt.savefig('figures/{}_BICs2-20'.format(model_folder), bbox_inches='tight')
```

```python
fig, axes = plt.subplots(1,2)
for m_id in ids:
    axes[0].plot(np.arange(2, 21), AICs[m_id], label=m_id)
    axes[1].plot(np.arange(3, 21), AICs[m_id][1:]-AICs[m_id][:-1], label=m_id)
fig.set_size_inches(12,6)
for ax in axes:
    ax.set_xlabel('Number of classes (K)')
    ax.set_xticks(np.arange(2,21,2))
axes[0].legend()
axes[0].set_title('AIC',fontsize=14,fontweight='bold')
axes[0].text(-0.08,1.0,'a)',transform=axes[0].transAxes,fontsize=16,fontweight='bold')
axes[1].set_title('$\Delta$AIC',fontsize=14,fontweight='bold')
axes[1].text(-0.08,1.0,'b)',transform=axes[1].transAxes,fontsize=16,fontweight='bold')
plt.savefig('figures/{}_AICs2-20'.format(model_folder), bbox_inches='tight')
```

```python
fig, axes = plt.subplots(1,2)
for m_id in ids:
    axes[0].plot(np.arange(2, 21), SILs[m_id], label=m_id)
    axes[1].plot(np.arange(3, 21), SILs[m_id][1:]-SILs[m_id][:-1], label=m_id)
fig.set_size_inches(12,6)
for ax in axes:
    ax.set_xlabel('Number of classes (K)')
    ax.set_xticks(np.arange(2,21,2))
axes[0].legend()
axes[0].set_title('SIL',fontsize=14,fontweight='bold')
axes[0].text(-0.08,1.0,'a)',transform=axes[0].transAxes,fontsize=16,fontweight='bold')
axes[1].set_title('$\Delta$SIL',fontsize=14,fontweight='bold')
axes[1].text(-0.08,1.0,'b)',transform=axes[1].transAxes,fontsize=16,fontweight='bold')
plt.savefig('figures/{}_SILs2-20'.format(model_folder), bbox_inches='tight')
```
